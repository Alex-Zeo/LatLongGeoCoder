```python
import csv
from opencage.geocoder import OpenCageGeocode

# API key from OpenCage Geocoder
key = 'yourapikey'
geocoder = OpenCageGeocode(key)

addresses = [ #Marta Stations:
    "6000 S Terminal Pkwy, Atlanta, GA 30337",
    "1255 West Peachtree St NE, Atlanta, GA 30309",
    "65 Joseph E. Lowery Blvd NW, Atlanta, GA 30314",
    "1335 Donald Lee Hollowell Pkwy NW, Atlanta, GA 30318",
    "4047 Peachtree Rd NE, Atlanta, GA 30319",
    "3360 Peachtree Rd NE, Atlanta, GA 30326",
    "5200 New Peachtree Rd, Chamblee, GA 30341",
    "435 West Peachtree St NW, Atlanta, GA 30308",
    "3800 Main St, College Park, GA 30337",
    "400 Church St, Decatur, GA 30030",
    "6000 New Peachtree Rd, Doraville, GA 30340",
    "1111 Hammond Dr NE, Atlanta, GA 30346",
    "2260 College Ave NE, Atlanta, GA 30307",
    "2848 East Main St, East Point, GA 30344",
    "1475 DeKalb Ave NE, Atlanta, GA 30307",
    "30 Alabama St SW, Atlanta, GA 30303",
    "225 Peachtree St SW, Atlanta, GA 30303",
    "170 Piedmont Ave SE, Atlanta, GA 30303",
    "70 Hamilton E Holmes Dr NW, Atlanta, GA 30311",
    "3901 Durham Park Rd, Stone Mountain, GA 30083",
    "1055 DeKalb Ave NE, Atlanta, GA 30307",
    "3350 Kensington Rd, Decatur, GA 30032",
    "377 Decatur St SE, Atlanta, GA 30312",
    "2020 Lee St SW, Atlanta, GA 30310",
    "955 East Paces Ferry Rd NE, Atlanta, GA 30326",
    "2424 Piedmont Rd NE, Atlanta, GA 30324",
    "5711 Peachtree Dunwoody Rd, Atlanta, GA 30342",
    "41 10th St NE, Atlanta, GA 30309",
    "713 West Peachtree St NW, Atlanta, GA 30308",
    "7010 Peachtree Dunwoody Rd, Sandy Springs, GA 30328",
    "1400 Lee St SW, Atlanta, GA 30310",
    "200 Peachtree St NE, Atlanta, GA 30303",
    "1101 Mount Vernon Hwy, Sandy Springs, GA 30338",
    "502 Rhodes St NW, Atlanta, GA 30314",
    "680 Lee St SW, Atlanta, GA 30310",
    "80 Anderson Ave NW, Atlanta, GA 30314"
]

# Prepare the CSV file to store the results
with open('geocode_results.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Address', 'Latitude', 'Longitude'])

    # Geocode each address
    for address in addresses:
        result = geocoder.geocode(address)
        if result:
            latitude = result[0]['geometry']['lat']
            longitude = result[0]['geometry']['lng']
            writer.writerow([address, latitude, longitude])
        else:
            writer.writerow([address, 'Not found', 'Not found'])

print("Geocoding complete, results saved to geocode_results.csv.")
```

```python
import folium

# Function to load geocoded results from CSV
def load_geocoded_results(filepath):
    data = []
    with open(filepath, mode='r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header
        for row in reader:
            if row[1] != 'Not found':
                data.append((row[0], float(row[1]), float(row[2])))
    return data

# Load the geocoded results
geocoded_data = load_geocoded_results('geocode_results.csv')

# Create a map centered around Atlanta, GA
map = folium.Map(location=[33.7490, -84.3880], zoom_start=11)

# Add markers for each geocoded address
for address, lat, lng in geocoded_data:
    folium.Marker([lat, lng], popup=address).add_to(map)

# Display the map
map
```
